```python
# location & url 구하기
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException

def get_loc_from_current_page(driver):
    location_list = []
    elements = driver.find_elements(By.CLASS_NAME, "list_body")
    for element in elements:
        locations = element.find_elements(By.CLASS_NAME, 'work_place')
        for location in locations:
            location_list.append(location.text)

            
    return location_list

def get_data_from_current_page(driver):
    elements = driver.find_elements(By.CLASS_NAME, "list_body")
    url_list = []
    for element in elements:
        strong_tags = element.find_elements(By.CLASS_NAME, 'str_tit')
        for strong_tag in strong_tags:
            url_list.append(strong_tag.get_attribute("href"))
    return url_list

# 해당 페이지로 이동하는 함수
def click_page_by_number(driver, page_number):
    try:
        xpath = f'//*[@id="default_list_wrap"]/div/button[{page_number}]'
        page_button = driver.find_element(By.XPATH, xpath)
        page_button.click()
        print(f"{page_number} 페이지 클릭")
        return True
    except NoSuchElementException:
        print(f"{page_number} 페이지를 찾을 수 없습니다.")
        return False

# 첫 번째 페이지부터 시작
location_list = []
url_list = []
# 현재 페이지에서 데이터 수집
current_loc = get_loc_from_current_page(driver)
location_list.extend(current_loc)

current_data = get_data_from_current_page(driver)
url_list.extend(current_data)




# 수집한 데이터 출력
print(f"현재 페이지 데이터 수집 완료. 수집된 데이터 수: {len(location_list)}")

# 1부터 9까지의 페이지 클릭하면서 데이터 수집
for page_number in range(1, 9):
    click_page_by_number(driver, page_number)
    # 현재 페이지에서 데이터 수집
    current_loc = get_loc_from_current_page(driver)
    location_list.extend(current_loc)
    
    # 현재 페이지에서 데이터 수집
    current_data = get_data_from_current_page(driver)
    url_list.extend(current_data)
    
    # 수집한 데이터 출력
    print(f"{page_number} 페이지 위치데이터 수집 완료. 수집된 위치데이터 수: {len(location_list)}")

# 수집한 URL 정보 출력
print(location_list)
print(url_list)
```

```python
driver = webdriver.Chrome()

company = []
title = []
career = []
education = []
Infodetails = []
tag = []
dday = []

urls = df['URL']
for url in urls:
    try:
        # 각 URL로 이동
        driver.get(url)
        
        # 페이지 로드 대기 (필요에 따라 조정)
        time.sleep(2)
        
        # 페이지 소스를 BeautifulSoup로 파싱
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        
        # 필요한 요소를 찾아 텍스트를 가져옵니다.
        try:
            company_list = soup.find("a", class_="company").text.strip()
        except:
            company_list = None
            
        try:
            title_select = soup.find("h1", class_="tit_job").text.strip()
        except:
            title_select = None
        
        try:
            career_select = soup.select_one('.jv_summary > div > div:nth-child(1) > dl:nth-child(1) > dd > strong').text.strip()
        except:
            career_select = None
        
        try:
            education_list = soup.select_one('.jv_summary > div > div:nth-child(1) > dl:nth-child(2) > dd > strong').text.strip()
        except:
            education_list = None
            
        try:
            info_list = soup.find("h2", "tit_template").text.strip()
        except:
            info_list = None
            
        try:
            tag_list = soup.find("ul", "scroll").text.strip().replace("#", ",").lstrip(',')#.split(',')
        except:
            tag_list = None
            
        try:
            dday_list = soup.find("dl", "info_period").text.strip()
        except:
            dday_list = None
           
            
        
        # 데이터를 리스트에 추가합니다.
        title.append(title_select)
        company.append(company_list)
        career.append(career_select)
        education.append(education_list)
        Infodetails.append(info_list)
        dday.append(dday_list)
        tag.append(tag_list)
    
    except Exception as e:
        print(f"Error processing URL {url}: {e}")

# 결과를 데이터프레임으로 변환합니다.
result_df = pd.DataFrame({
    
    'Location': df['Location'],
    'Company': company,
    'Title': title,
    'Career': career,
    'Edu': education,
    'URL': df['URL'],
    'InfoDetails': Infodetails, 
    'Tag': tag,
    'D_Day': dday
})

# 결과 출력
print(result_df)
```

```python

df = pd.read_csv("C:\\Users\Administrator\AICC 수업자료\\프로젝트.csv", encoding='utf-8')
df = df.fillna('')
conn = pymysql.connect(host='127.0.0.1', user='user', password='user1234', db='dbmaria', charset='utf8')
cur = conn.cursor()

file_name = "프로젝트"
cur.execute('''
    CREATE TABLE IF NOT EXISTS project(
        company VARCHAR(255),
        title VARCHAR(255),
        career VARCHAR(255),
        edu VARCHAR(255),
        urls VARCHAR(255),
        infodetails VARCHAR(255),
        tag TEXT,
        d_day VARCHAR(255),
        location VARCHAR(255)
    )
''')

# 데이터프레임의 각 행을 삽입
for index, row in df.iterrows():
    cur.execute("""
        INSERT INTO project (
           company,
           title,
           career,
           edu,
           urls,
           infodetails,
           tag,
           d_day,
           location
        ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
    """, (
        row['Company'],
        row['Title'],
        row['Career'],
        row['Edu'],
        row['URL'],
        row['InfoDetails'],
        row['Tag'],
        row['D_Day'],
        row['Location']
    ))


# 변경사항 커밋
conn.commit()

# 연결 종료
conn.close()
```
