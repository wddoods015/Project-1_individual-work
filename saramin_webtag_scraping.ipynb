```python
import requests
from bs4 import BeautifulSoup
import re
from konlpy.tag import Okt
from collections import Counter
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
tag_df = pd.read_csv('C:\\Users\\EZEN\\강의자료\\project-2\\web.csv', encoding='cp949')
```

```python
Tag = tag_df['Tag'] 
Tag.info()
```

```python
tag = Tag.tolist()
```

```python
words = ", ".join(str(lst) for lst in tag)
words
```

```python
word_list = words.split(',')

# 각 단어의 빈도수 계산
word_count = Counter(word_list)
```

```python
word_count
```

```python
web_stack = pd.DataFrame(list(word_count.items()), columns=['Word', 'Frequency'])
web_stack 
```

```python
web_stack.to_csv('web_stack.csv', index=False, encoding='cp949')
```

```python
words_tagger = Okt() # 명사 추출
nouns = words_tagger.nouns(words)
count = Counter(nouns) # 각 명사의 빈도수 계산
```

```python
count  
```

```python
# 빈도수 특정 명사의 갯수 지정
ranked_tags = count.most_common(100) # 빈도순 n개 단어
```
